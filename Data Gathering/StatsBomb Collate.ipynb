```python
import csv
import pandas as pd
import os
route_Dir = r'C:\Users\jacks\OneDrive\Documents\Work\Football\Datasets\StatsBomb\2020-21\\'
```

```python
def normalise_Data_File(data_file,column_headers):
    df = pd.read_csv(route_Dir+data_file, encoding="ISO-8859-1")
    df = df.iloc[1: , :]
    df.columns = column_headers
    del df['Matches']
    print(data_file+' Complete')
    return df
```

```python
standardStats_column_headers=['Rk','Player','Nation','Pos','Squad','Age','Born',
                            'MP','Starts','Min','90s',
                            'Gls', 'Ast','G-PK','PK','PKatt','CrdY','CrdR',
                            'Gls/90','Ast/90','G+A/90','G-PK/90','G+A-PK/90',
                            'xG','npxG','xA','npxG+xA',
                            'xG/90','xA/90','xG+xA/90', 'npxG/90','npxG+xA/90',
                            'Matches'
                ]

df_standardStats = normalise_Data_File('Standard Stats.csv',standardStats_column_headers)

goalkeepingStats_column_headers = ['Rk','Player','Nation','Pos','Squad','Age','Born',
                                    'MP','Starts','Min','90s',
                                    'GA','GA90','SoTA','Saves','Save%',
                                    'W','D','L',
                                    'CS','CS%','PKatt','PKA','PKsv','PKm','Save%PK',
                                    'Matches'
]

df_goalkeepingStats = normalise_Data_File('Goalkeeping Stats.csv',goalkeepingStats_column_headers)

AdvGoalkeeping_column_headers = ['Rk','Player','Nation','Pos','Squad','Age','Born',
                                '90s','GA','PKA','FK','CK','OG',
                                'PSxG','PSxG/SoT','PSxG+/-','/90',
                                'Cmp','Att','Cmp%',
                                'Att','Thr','Launch%','AvgLen',
                                'AttGK','Launch%GK','AvgLenGK',
                                'Opp','Stp','Stp%',
                                '#OPA','#OPA/90','AvgDist',
                                'Matches']

df_AdvGoalkeepingStats = normalise_Data_File('Advanced Goalkeeping Stats.csv',AdvGoalkeeping_column_headers)

shooting_column_headers = [ 'Rk','Player','Nation','Pos','Squad','Age','Born',
                            '90s','Gls','Sh','SoT','SoT%','Sh/90','SoT/90','G/Sh','G/SoT','Dist','FK',
                            'PK','PKatt',
                            'xG','npxG','npxG/Sh','G-xG','np:G-xG',
                            'Matches'

]

df_shootingStats = normalise_Data_File('Shooting Stats.csv',shooting_column_headers)

passing_column_headers=['Rk','Player','Nation','Pos','Squad','Age','Born',
                        '90s','TotCmp','TotAtt','TotCmp%','TotDist','PrgDist',
                        'SrtCmp','SrtAtt','SrtCmp%',
                        'MedCmp','MedAtt','MedCmp%',
                        'LngCmp','LngAtt','LngCmp%',
                        'Ast','xA','A-xA','KP','P3rd','PPA','CrsPA','Prog',
                        'Matches'
]

df_passingStats = normalise_Data_File('Passing Stats.csv',passing_column_headers)

passingTypes_column_headers=['Rk','Player','Nation','Pos','Squad','Age','Born',
                            '90s','Att','Live','Dead','FK','TB','Press','Sw','Crs','CK',
                            'InCK','OutCK','StrCK',
                            'Ground','Low','High',
                            'LeftBP','RightBP','HeadBP','TIBP','OtherBP',
                            'Cmp','OffOC','OutOC','IntOC','BlocksOC',
                            'Matches'

]

df_passTypes = normalise_Data_File('Passing Types Stats.csv',passingTypes_column_headers)

GoalandShotCreation_column_headers=['Rk','Player','Nation','Pos','Squad','Age','Born',
                            '90s','SCA','SCA90','PassLive','PassDead',
                            'Drib','Sh2sh','Fld','Def','GCA','GCA90','PassLiveG','PassDeadG','DribG','ShG','FldG','DefG',
                            'Matches'
]

df_GoalandShotCreation = normalise_Data_File('Goal and Shot Creating Stats.csv',GoalandShotCreation_column_headers)

DefensiveActions_column_headers=['Rk','Player','Nation','Pos','Squad','Age','Born',
                            '90s','Tkl','TklW','Def 3rd','Mid 3rd','Att 3rd',
                            'TklDbl','AttDbl','Tkl%Dbl','PastDbl',
                            'Press','PressSucc','Press%','PressDef3rd','PressMid3rd','PressAtt3rd',
                            'Blocks','ShBlocks','ShSvBlocks','PassBlocks',
                            'Int','Tkl_Int','Clr','Err',
                            'Matches'
]

df_DefensiveActions = normalise_Data_File('Defensive Actions Stats.csv',DefensiveActions_column_headers)

PlayingTime_column_headers=['Rk','Player','Nation','Pos','Squad','Age','Born',
                            'MP','Min','Mn/MP','Min%','90s','Starts',
                            'Mn/Start','Compl','Subs','Mn/Sub','unSub',
                            'PPM','onG','onGA','+/-','+/-90','On-Off','onxG','onxGA','xG+/-','xG+/-90','On-Off',
                            'Matches'
]

df_PlayingTime = normalise_Data_File('Playing Time Stats.csv',PlayingTime_column_headers)

Miscellaneous_column_headers=['Rk','Player','Nation','Pos','Squad','Age','Born',
                            '90s','CrdY','CrdR','2CrdY','Fls','Fld','Off','Crs','Int','TklW',
                            'Pkwon','PKcon','OG','Recov','Won','Lost','Won%',
                            'Matches'
]

df_Miscellaneous = normalise_Data_File('Miscellaneous Stats.csv',Miscellaneous_column_headers)
```

```python
#print(len(df_AdvGoalkeepingStats))
#print(len(df_goalkeepingStats))
#print(len(df_PlayingTime))

dfs_toMerge = [df_shootingStats,df_passTypes,df_passingStats,df_GoalandShotCreation,df_DefensiveActions,df_Miscellaneous]
df_PlayerStats20_21 = df_standardStats

for df in dfs_toMerge:
    del df['Rk'],df['Player'], df['Nation'],df['Pos'],df['Squad'],df['Age'],df['Born'],df['90s']
    df_PlayerStats20_21 = pd.concat([df_PlayerStats20_21, df], axis=1, join="inner")

df_PlayerStats20_21.to_csv(route_Dir+'Combined Data.csv',index=False)
df_PlayerStats20_21
```

Collate combined datasets for each season into a single dataframe

```python
folderDir = r'C:\Users\jacks\OneDrive\Documents\Work\Football\Datasets\StatsBomb\\'
#folders = os.listdir(folderDir) #manual to ensure order is probably best practice
folders = ['2020-21','2017-18']

df_combined = pd.read_csv(folderDir+folders[0]+'\\Combined Data.csv', encoding="ISO-8859-1").add_prefix('y'+folders[0]+'_')
df_combined.columns.values[1] = "Player"
df_combined = df_combined.set_index('Player')

for folder in folders[1:]:
    df = pd.read_csv(folderDir+folder+'\\Combined Data.csv', encoding="ISO-8859-1").add_prefix('y'+folder+'_')
    df.columns.values[1] = "Player"
    df = df.set_index('Player')
    df_combined = df_combined.join(df)

df_combined

```

Create Dataframe containing player data from just the years 17/18 and 20/21.

Filter this further to Just those in the forward positions

```python
roles=['FW','FWDF','FWMF','MFFW']
df_17_20_FW = df_combined[((df_combined['y2017-18_Pos'].notnull()) & (df_combined['y2020-21_Pos'].isin(roles)))]
df_17_20_FW
```

Label the Rows.
At the moment it is very simple: 8 goals or over is a 1.

This is a first attempt, so is simple by design

```python
pd.options.mode.chained_assignment = None  # default='warn'
def calc_label(row):
    if row['y2020-21_Gls'] < 8:
        value = 0
    else:
        value = 1
    return value

df_17_20_FW['label'] = df_17_20_FW.apply(calc_label, axis=1)
df_17_20_FW['label'].value_counts()

```

```python
#df_17_20_FW = df_17_20_FW.iloc[: , 150:]
df_17_20_FW = df_17_20_FW[df_17_20_FW.columns[-150:]]
df_17_20_FW
```

```python

df_17_20_FW = df_17_20_FW.drop(['y2017-18_Nation',
                                'y2017-18_Pos',
                                'y2017-18_Squad']
                                ,1)
df_17_20_FW
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df_17_20_FW), columns= df_17_20_FW.columns)

df_normalized.dropna()
df_normalized
```

```python
from sklearn.model_selection import train_test_split
y = df_17_20_FW[['label']]
X = df_normalized.drop(['label'],1)


print('Length of x: ',len(X))
print('Length of y: ',len(y))

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

X_test.fillna(X_train.mean(), inplace=True)
X_train.fillna(X_train.mean(), inplace=True)


print('Length of X_train: ',len(X_train))
print('Length of X_test: ',len(X_test))
print('Length of y_train: ',len(y_train))
print('Length of y_test: ',len(y_test))
```

SVM Classifier

```python
from sklearn.svm import SVC
svclassifier = SVC(kernel='linear')
svclassifier.fit(X_train, y_train)

y_pred = svclassifier.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
RFclassifier = RandomForestClassifier()
RFclassifier.fit(X_train, y_train)

y_pred = RFclassifier.predict(X_test)


print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))

out = y_test
out['pred'] = y_pred
out
```

Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=2)
neigh.fit(X_train, y_train)
```

```python
y_pred = neigh.predict(X_test)
```
