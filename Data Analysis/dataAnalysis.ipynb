```python
df_collated = pd.read_csv(r'F:\Football Analytics\StatsBomb\collatedData17-21.csv',low_memory=False)
df_collated.head()
```

Description

```python
df_avgGoalperAge17_18 = df_collated[['y17-18_Age', 'y17-18_Gls']].groupby(['y17-18_Age'], as_index=False).mean().sort_values(by='y17-18_Age', ascending=False)
df_avgGoalperAge17_18
```

```python
df_avgGoalperAge21_22 = df_collated[['y21-22_Age', 'y21-22_Gls']].groupby(['y21-22_Age'], as_index=False).mean().sort_values(by='y21-22_Age', ascending=False)
df_avgGoalperAge21_22
```

```python
plt.bar(df_avgGoalperAge21_22['y21-22_Age'],df_avgGoalperAge21_22['y21-22_Gls'])
plt.xlabel('Age')
plt.ylabel('Avg #Goals')
plt.title('Average number of goals per player age 21-22')
plt.show
```

```python
plt.rcParams['figure.figsize'] = (8, 5)

fig, ax = plt.subplots()
bars = ax.bar(df_avgGoalperAge21_22['y21-22_Age'],df_avgGoalperAge21_22['y21-22_Gls'])

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('#DDDDDD')
ax.tick_params(bottom=False, left=False)
ax.set_axisbelow(True)
ax.yaxis.grid(True, color='#EEEEEE')
ax.xaxis.grid(False)

bar_color = bars[0].get_facecolor()

ax.set_xlabel('Age', labelpad=15, color='#333333')
ax.set_ylabel('Average Goals Scored', labelpad=15, color='#333333')
ax.set_title('Average Number of Goals Scored by Age (21-22)', pad=15, color='#333333',
             weight='bold')

fig.tight_layout()
```

```python
plt.rcParams['figure.figsize'] = (8, 5)

fig, ax = plt.subplots()
bars = ax.bar(df_avgGoalperAge21_22['y21-22_Age'],df_avgGoalperAge21_22['y21-22_Gls'])

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('#DDDDDD')
ax.tick_params(bottom=False, left=False)
ax.set_axisbelow(True)
ax.yaxis.grid(True, color='#EEEEEE')
ax.xaxis.grid(False)

bar_color = bars[0].get_facecolor()

ax.set_xlabel('Age', labelpad=15, color='#333333')
ax.set_ylabel('Average Goals Scored', labelpad=15, color='#333333')
ax.set_title('Average Number of Goals Scored by Age (21-22)', pad=15, color='#333333',
             weight='bold')

fig.tight_layout()
```

```python
import numpy as np

labels = df_avgGoalperAge17_18['y17-18_Age']
width = 0.35  # the width of the bars

x = np.arange(len(labels))
fig, ax = plt.subplots()
rects1 = ax.bar(labels - width/2, df_avgGoalperAge17_18['y17-18_Gls'], width, label='17-18')
rects2 = ax.bar(labels + width/2, df_avgGoalperAge21_22['y21-22_Gls'], width, label='21-22')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
#ax.set_xticks(x, labels)
ax.legend()
ax.set_xlabel('Age', labelpad=15, color='#333333')
ax.set_ylabel('Average Goals Scored', labelpad=15, color='#333333')
ax.set_title(r"Average Number of Goals Scored in the Premier League by Age""\n"r"(Seasons: 17-18 & 21-22)", pad=15, color='#333333',
             weight='bold')


# 
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('#DDDDDD')
ax.tick_params(bottom=False, left=False)
ax.set_axisbelow(True)
ax.yaxis.grid(True, color='#EEEEEE')
ax.xaxis.grid(False)

fig.tight_layout()
plt.show()
```
